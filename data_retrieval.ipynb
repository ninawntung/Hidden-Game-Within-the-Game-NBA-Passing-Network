```python
from nba_api.stats.static import teams

# Get all NBA teams
nba_teams = teams.get_teams()

# Convert to DataFrame for easier viewing
import pandas as pd
teams_df = pd.DataFrame(nba_teams)

# Display all teams
teams_df[['id', 'full_name', 'abbreviation', 'city', 'nickname']]
```

```python
from nba_api.stats.endpoints import playerdashptpass, commonteamroster
import pandas as pd
import time

# Step 1: Get all players on the Pistons roster for 2024-25 season
pistons_roster = commonteamroster.CommonTeamRoster(
    team_id=1610612760,  # Change team ID
    season='2024-25'
)

# Get the roster dataframe
roster_df = pistons_roster.common_team_roster.get_data_frame()

# Step 2: Loop through each player and get their passing data
all_passing_data = []

for index, player in roster_df.iterrows():
    player_id = player['PLAYER_ID']
    player_name = player['PLAYER']
    
    print(f"Getting passing data for {player_name} (ID: {player_id})...")
    
    try:
        # Get passing data for this player
        passing = playerdashptpass.PlayerDashPtPass(
            player_id=player_id,
            season='2024-25',
            team_id=1610612760  # Change team ID
        )
        
        # Get the passes made dataframe
        passes_made_df = passing.passes_made.get_data_frame()
        
        # Add passer name for reference
        passes_made_df['PASSER_NAME'] = player_name
        
        # Append to our collection
        all_passing_data.append(passes_made_df)
        
        # Be nice to the API - wait between requests
        time.sleep(0.6)  # 600ms delay to avoid rate limiting
        
    except Exception as e:
        print(f"Error getting data for {player_name}: {e}")
        continue

# Step 3: Combine all data into one dataframe
if all_passing_data:
    complete_passing_df = pd.concat(all_passing_data, ignore_index=True)
    print(f"\nTotal passing connections: {len(complete_passing_df)}")
    print(complete_passing_df.head(20))
else:
    print("No passing data collected")
```

```python
complete_passing_df
```

```python
# Filter and rename the DataFrame columns
filtered_df = complete_passing_df[['PLAYER_NAME_LAST_FIRST', 'PASS_TO', 'TEAM_ABBREVIATION', 'PASS', 'AST']]

# Function to reformat player name (Last, First -> First Last)
def reformat_name(name):
    if ',' in name:
        last, first = name.split(',')
        return f"{first.strip()} {last.strip()}"
    return name

# Reformat player names
filtered_df['PLAYER_NAME_LAST_FIRST'] = filtered_df['PLAYER_NAME_LAST_FIRST'].apply(reformat_name)
filtered_df['PASS_TO'] = filtered_df['PASS_TO'].apply(reformat_name)

# Rename columns to lowercase
filtered_df.columns = ['player_name', 'pass_to', 'team', 'pass', 'ast']

# Display the filtered DataFrame with new column names and reformatted names
print("Filtered passing data with reformatted names:")
filtered_df
```

```python
# save into the folder passing_data 
filtered_df.to_csv('passing_data/Thunder_improve/Thunder_passing_data_2024_25.csv', index=False)
```
